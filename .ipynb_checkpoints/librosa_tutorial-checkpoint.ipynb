```python
import glob
import os
import librosa
from librosa import display as lbdisplay
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from matplotlib.pyplot import specgram
import IPython.display as ipd
```

```python
def load_sound_files(file_paths):
    raw_sounds = []
    for fp in file_paths:
        # X : numpy array
        X,sr = librosa.load(fp)
        raw_sounds.append(X)
    return raw_sounds

def plot_waves(sound_names,raw_sounds):
    i = 1
    fig = plt.figure(figsize=(25,60), dpi = 900)
    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(10,1,i)
        lbdisplay.waveplot(np.array(f),sr=22050)
        plt.title(n.title())
        i += 1
    plt.suptitle("Figure 1: Waveplot",x=0.5, y=0.915,fontsize=18)
    plt.show()
    
def plot_specgram(sound_names,raw_sounds):
    i = 1
    fig = plt.figure(figsize=(25,60), dpi = 900)
    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(10,1,i)
        specgram(np.array(f), Fs=22050)
        plt.title(n.title())
        i += 1
    plt.suptitle("Figure 2: Spectrogram",x=0.5, y=0.915,fontsize=18)
    plt.show()

def plot_log_power_specgram(sound_names,raw_sounds):
    i = 1
    fig = plt.figure(figsize=(25,60), dpi = 900)
    for n,f in zip(sound_names,raw_sounds):
        plt.subplot(10,1,i)
        # https://librosa.github.io/librosa/generated/librosa.core.stft.html
        # stftの絶対値(実部)がamplitude
        # power spectrogram : amplitudeの2乗
        D = librosa.power_to_db(np.abs(librosa.stft(f))**2)
        lbdisplay.specshow(D,x_axis='time' ,y_axis='log')
        plt.title(n.title())
        i += 1
    plt.suptitle("Figure 3: Log power spectrogram",x=0.5, y=0.915,fontsize=18)
    plt.show()
```

```python
sound_file_paths = ["data/" + f for f in ["Hats/Hat_1.wav", "Kicks/Kick_1.wav", "Snares/Snare_1.wav"]]
sound_names = ["Hat","Kick","Snare"]
ipd.Audio(sound_file_paths[0])
```

```python
raw_sounds = load_sound_files(sound_file_paths)

plot_waves(sound_names,raw_sounds)
```

```python
plot_specgram(sound_names,raw_sounds)
```

特徴量1. mfcc: Mel-frequency cepstral coefficients
http://practicalcryptography.com/miscellaneous/machine-learning/guide-mel-frequency-cepstral-coefficients-mfccs/
https://qiita.com/martin-d28jp-love/items/34161f2facb80edd999f

1980年代に提案された音声認識でよく使われる特徴量。

Mel尺度 : 人間の音高の知覚尺度。メル尺度の差が同じなら、音高の差が同じに感じる。周波数に対してlog

MFCCの計算方法
1. STFT
2. メル尺度でとった三角フィルターをかけて、その中でamplitudeを足し合わせてlogをとる→フィルターの数ぶんの次元のベクトル
3. 2でできたものを離散コサイン変換する


```python
X, sample_rate = librosa.load(sound_file_paths[0])
print(sample_rate)
mfccs = librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40)
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
# 平均を取る意味？
mfccs = np.mean(mfccs.T,axis=0)
print(mfccs)
```
